# Basics

## Imports etc.

```python
import pandas as pd
import numpy as np
import os
import plotnine as gg
gg.theme_set(gg.theme_classic)
```

## Directories etc.

```python
data_save_path = "C:/Users/maria/MEGAsync/SLCNdata/Meta/all_data"
plot_save_path = "C:/Users/maria/MEGAsync/SLCN/MetaSLCNPlots"
if not os.path.exists(plot_save_path):
    os.makedirs(plot_save_path)
```

## Get data

```python
all_data = pd.read_csv(os.path.join(data_save_path, 'all_data_raw.csv'), index_col=0)
```

# RLWM

```python
gg.options.figure_size = (3, 3)
g = (gg.ggplot(all_data, gg.aes('age_group_', 'rl_ACC_ns_slope', group=1))  # , color='sex', group='sex'
     + gg.stat_summary()
     + gg.stat_summary(geom='line', linetype='dotted')
     + gg.theme(axis_text_x=gg.element_text(rotation=90, hjust=0.5), legend_position=(0.75, 0.85))
     + gg.labs(x='', y='Set size effect', color='')
    )
print(g)
g.save(os.path.join(plot_save_path, '10_SummaryRLWM.svg'))
g.save(os.path.join(plot_save_path, '10_SummaryRLWM.png'))
```

```python
g = (gg.ggplot(all_data, gg.aes('PreciseYrs', 'rl_ACC_ns_slope'))
     + gg.geom_point()
     + gg.stat_smooth(method='lm', formula='y ~ x + np.square(x)')
    )
g
```

## Butterfly

```python
bf_dat = all_data[['ID', 'age_group_', 'bf_ACC_early', 'bf_ACC_mid', 'bf_ACC_late']]  # [c for c in all_data.columns if 'bf_' in c]
bf_long = bf_dat.melt(id_vars=['ID', 'age_group_'], value_name='ACC', var_name='block') \
                .replace({'bf_ACC_early': 1, 'bf_ACC_mid': 2, 'bf_ACC_late': 3})
bf_long
```

```python
g = (gg.ggplot(bf_long, gg.aes('age_group_', 'ACC', color='factor(block)', group='factor(block)'))
     + gg.stat_summary()
     + gg.stat_summary(geom='line', linetype='dotted')
     + gg.labs(x='', y='Accuracy', color='Block')
     + gg.scales.scale_color_brewer(type='seq', palette=6)
     + gg.theme(axis_text_x=gg.element_text(rotation=90, hjust=0.5))
    )
print(g)
g.save(os.path.join(plot_save_path, '10_SummaryBF.png'))
g.save(os.path.join(plot_save_path, '10_SummaryBF.svg'))
```

```python
g = (gg.ggplot(all_data, gg.aes('age_group_', 'bf_ACC', group=1))
     + gg.stat_summary()
     + gg.stat_summary(geom='line', linetype='dotted')
     + gg.labs(x='')
     + gg.theme(axis_text_x=gg.element_text(rotation=90, hjust=0.5))
    )
print(g)
```
