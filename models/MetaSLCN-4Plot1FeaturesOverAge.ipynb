# Imports

```python
import numpy as np
import os
import pandas as pd
import plotnine as gg
gg.theme_set(gg.theme_classic)
```

## Directories etc.

```python
data_save_path = "C:/Users/maria/MEGAsync/SLCNdata/Meta/all_data"
plot_save_path = "C:/Users/maria/MEGAsync/SLCN/MetaSLCNPlots"
if not os.path.exists(plot_save_path):
    os.makedirs(plot_save_path)
```

# Get data

```python
all_data = pd.read_csv(os.path.join(data_save_path, 'all_data.csv'), index_col=0)
all_data_z = pd.read_csv(os.path.join(data_save_path, 'all_data_z.csv'), index_col=0)
all_data_long = pd.read_csv(os.path.join(data_save_path, 'all_data_long.csv'), index_col=0)
all_data_long_z = pd.read_csv(os.path.join(data_save_path, 'all_data_long_z.csv'), index_col=0)
```

## Features over age

```python
# Each feature over age. PDS, T
gg.options.figure_size = (20, 20)
g = (gg.ggplot(all_data_long, gg.aes('PreciseYrs', 'value', color='factor(Gender)'))
 + gg.geom_point(size=0.5, alpha=0.5)
 + gg.geom_smooth()
 + gg.labs(y='')
 + gg.facet_wrap('~ feature', scales='free')
)
g.save(os.path.join(plot_save_path, '1_FeaturesOverAge.png'))

g_pds = g + gg.aes('PDS')
g_pds.save(os.path.join(plot_save_path, '1_FeaturesOverPDS.png'))
g_t = g + gg.aes('meanT')
g_t.save(os.path.join(plot_save_path, '1_FeaturesOverT.png'))

print(g)
```

```python
# Each feature over age group
g2 = (gg.ggplot(all_data_long, gg.aes('age_group', 'value', fill='factor(age_group)'))
 + gg.stat_summary(geom='bar')
 + gg.scale_fill_brewer()
 + gg.facet_wrap('~ feature', scales='free')
)
g2.save(os.path.join(plot_save_path, '1_FeaturesOverAgeGroups.png'))

print(g2)
```

```python
# Feature groups over age
gg.options.figure_size = (10, 10)

g = (gg.ggplot(all_data_long, gg.aes('PreciseYrs', 'value', color='feature'))
     + gg.labs(y='', color='')
     + gg.geom_smooth()
     + gg.facet_wrap('~ category', scales='free')
)
g.save(os.path.join(plot_save_path, '1_FeaturesOverAge_Grouped.png'))
print(g)

g_z = g
g_z.data = all_data_long_z
g_z.save(os.path.join(plot_save_path, '1_FeaturesOverAge_Grouped_z.png'))

print(g_z)
```
